```python
import os

IMAGES_PATH = "Train/Images/"
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Train/train.csv')
df.head()

labels_array = df.values
```

```python
img_2_label = {}
for i in range(labels_array.shape[0]):
    img_id = labels_array[i][0]
    label = labels_array[i][1]
    img_2_label[img_id] = label
    
```

```python
import matplotlib.pyplot as plt
import cv2

def readImg(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    #plt.imshow(img)
    return img
```

```python
img_2_label['296.jpg']
```

```python
def readAllImgs():
    
    X = []
    files = os.listdir(IMAGES_PATH)
    for f in files:
        path =  IMAGES_PATH + f
        #print(path)
        img = readImg(path)
        img = cv2.resize(img,(100,100))
        X.append(img)
    return X
        
X = readAllImgs()
    
```

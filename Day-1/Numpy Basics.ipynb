```python
a = np.array([[1,2],
              [3,4]])

b = np.array([[1,0],
             [3,2]])
print(a*b)
```

```python
# Matrix Multipcation
c = np.dot(a,b)
print(c)
```

```python
c = c.T
```

```python
# Hstack & Vstack
np.hstack((a,b))
```

```python
m = np.vstack((a,b,c,c))
print(m)
```

```python
## OpenCV
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
```

```python
cropped_img_2 = img[420:820, 1300: , : ]
plt.imshow(cropped_img_2)
plt.show()
```

```python
cropped_img = img[420:820, 450:950, : ]
plt.imshow(cropped_img)
plt.show()
```
